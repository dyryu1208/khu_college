```python
lime_df = pd.read_csv('/content/drive/MyDrive/seminar/lime_word_list.csv')
print(lime_df.shape)
lime_df.head()
```

```python
useful = lime_df[lime_df['model_predict'] == 1]
print(useful.shape)
useful = useful.reset_index(drop=True)
useful.head()
```

```python
def regex(x):
  cleaned_x = re.sub('[^a-z0-9.,-]',' ', x)

  return cleaned_x

useful['cleaned_word'] = useful['word_list'].apply(lambda x: regex(x))
useful.head()
```

```python
preprocessed_text = []
for i in range(len(useful)):
  word_list = useful['cleaned_word'][i]
  text_or_num = word_list.split(',')
  text = []
  for j in range(len(text_or_num)):

    if j % 2 == 0:
      if float(text_or_num[j+1]) > 0:
        text.append(text_or_num[j].strip())
      else:
        pass
    else:
      pass
  preprocessed_text.append(text)
```

```python
useful['only_word'] = preprocessed_text
```

```python
useful.head()
```

```python
result = []
for i in range(len(useful)):
  result.extend(useful['only_word'][i])

len(result)
```

```python
word_counter = Counter(result)
print(word_counter)
```

```python
word = []
count = []
for k, v in dict(word_counter).items():
  word.append(k)
  count.append(v)

dictionary = {'단어명' : word, '등장 횟수' : count}
word_df = pd.DataFrame(dictionary)
word_df = word_df.sort_values(by='등장 횟수', ascending=False)
word_df.head()
```

```python
word_df.to_csv('lime 단어 정리.csv', encoding='utf-8-sig', index=False)
```
